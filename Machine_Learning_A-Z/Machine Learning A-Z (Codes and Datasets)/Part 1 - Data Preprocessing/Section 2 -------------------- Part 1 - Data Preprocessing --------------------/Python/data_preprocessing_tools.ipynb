## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv("Data.csv")

# Creating the matrix of features (X) and the dependent variable vector (y)
x = dataset.iloc[:, :-1].values # We take all the rows first, and then all columns minus the last one.
y = dataset.iloc[:, -1].values # Same as before, but only the last column.
```

```python
print(x)
```

```python
print(y)
```

## Taking care of missing data

```python
from sklearn.impute import SimpleImputer

# Replace missing values by the average of their column
imputer = SimpleImputer(missing_values=np.nan, strategy="mean")
imputer.fit(x[:, 1:3])
x[:, 1:3] = imputer.transform(x[:, 1:3])
```
